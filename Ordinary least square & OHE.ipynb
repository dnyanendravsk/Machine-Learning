Hypothesis:

MKT is a good predictor for PROFIT(Null hypothesis)  
    MKT is not a good predictor for PROFIT(Alternate Hypothesis)  

# Model OLS  (Ordinary Least Square)

#Feature selection  (Selection of important columns in the model)


# Forward selection --> add opne predictor at a time from model 


# Backward  elemination -->remove one predictor at a time from model 

```python
Y = A[["PROFIT"]]
X = A[["RND","MKT"]]

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.2,random_state=21)

from statsmodels.api import OLS,add_constant
xconst = add_constant(xtrain)
ols = OLS(ytrain,xconst)
model = ols.fit()
model.summary()

# This adds a constant column with 1 value, this is done to get b0(intercept)

# pvalue of a column idealy should be lesser than 0.05, means its a good predictor
```

```python
model.rsquared_adj
```

```python
A = pd.read_csv("C:/Users/Pawan/Downloads//Cars93.csv")
A.head()
```

```python
Y = A[["Width"]]

from warnings import filterwarnings
filterwarnings("ignore")

import seaborn as sb
import matplotlib.pyplot as plt
for i in A.columns:
    if(A[i].dtypes == "object"):
        sb.boxplot(A[i],A.Width)
        plt.show()
    else:
        sb.scatterplot(A[i],A.Width)
        plt.show()
```

```python
A.corr()["Width"].sort_values()
```

```python
A.columns
```

Columns like id, Make,Model,Manufacturer are containing huge number of unque values.
These columns are not statisttically important, hence we will drop them.

```python
X = A.drop(labels=["id","Width","Max.Price","Price","Rear.seat.room","Luggage.room","Make","Model","Manufacturer"],axis=1)
```

```python
cat = []
con = []
for i in X.columns:
    if(X[i].dtypes == "object"):
        cat.append(i)
    else:
        con.append(i)
```

```python
cat
```

```python
con
```

```python
X1 = pd.get_dummies(X[cat])
```

```python
X1.head()
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()

ss.fit_transform(X[con])
```

```python
X2 = pd.DataFrame(ss.fit_transform(X[con]),columns=con)
```

```python
X2.head()
```

```python
Xnew = X2.join(X1)
```

```python
Xnew.head()
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(Xnew,Y,test_size=0.2,random_state=21)
from statsmodels.api import OLS,add_constant
xconst = add_constant(xtrain)
ols = OLS(ytrain,xconst)
model = ols.fit()
model.summary()
```

```python
Q = pd.DataFrame(model.pvalues,columns=["pval"])
x = Q.pval.max()
col_to_drop = Q[Q.pval == x].index[0]
```

```python
col_to_drop
```

```python
Xnew = Xnew.drop(labels=[col_to_drop],axis=1)
xtrain,xtest,ytrain,ytest=train_test_split(Xnew,Y,test_size=0.2,random_state=21)
from statsmodels.api import OLS,add_constant
xconst = add_constant(xtrain)
ols = OLS(ytrain,xconst)
model = ols.fit()
model.summary()
```

```python
Q = pd.DataFrame(model.pvalues,columns=["pval"])
x = Q.pval.max()
col_to_drop = Q[Q.pval == x].index[0]
```

```python
Xnew = Xnew.drop(labels=[col_to_drop],axis=1)
xtrain,xtest,ytrain,ytest=train_test_split(Xnew,Y,test_size=0.2,random_state=21)
from statsmodels.api import OLS,add_constant
xconst = add_constant(xtrain)
ols = OLS(ytrain,xconst)
model = ols.fit()
model.summary()
```

```python
Q = pd.DataFrame(model.pvalues,columns=["pval"]) # create df of all pvalues
x = Q.pval.max() # find out highest p value
col_to_drop = Q[Q.pval == x].index[0]  # find out that this highest value is for which column

#Drop that column

Xnew = Xnew.drop(labels=[col_to_drop],axis=1)  # Dropping the column that has highest pval
xtrain,xtest,ytrain,ytest=train_test_split(Xnew,Y,test_size=0.2,random_state=21)
from statsmodels.api import OLS,add_constant
xconst = add_constant(xtrain)
ols = OLS(ytrain,xconst)
model = ols.fit()
model.summary()
```
